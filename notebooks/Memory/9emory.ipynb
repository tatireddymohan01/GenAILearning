## Buffer Memory

Buffer Memory stores the **full conversation history** for the current session.
Every message (human + AI) is kept in memory and sent back to the LLM on each turn.

```python
from langchain_openai import ChatOpenAI
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

llm   = ChatOpenAI(model='gpt-4o-mini')
store = {}  # session_id → InMemoryChatMessageHistory

def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]

chain = RunnableWithMessageHistory(llm, get_session_history)

cfg = {'configurable': {'session_id': 'user_1'}}

r1 = chain.invoke('Hi, my name is Mohan', config=cfg)
print('Bot:', r1.content)

r2 = chain.invoke('I work in Data Analytics', config=cfg)
print('Bot:', r2.content)

r3 = chain.invoke('What is my name?', config=cfg)
print('Bot:', r3.content)
```

```python
r4 = chain.invoke('What is my name? What do you know about me?', config=cfg)
print('Bot:', r4.content)
```

```python
# View stored memory
for msg in store['user_1'].messages:
    print(f'[{msg.type}] {msg.content}')
```
