## Concept 1 — Basic PromptTemplate

### What is it?
The simplest form of prompting. A string template with `{variables}` that get filled in at runtime.

### Pipeline
```
Template string → fill {variables} → final prompt string → LLM → Answer
```

### Limitation (what Concept 2 fixes)
No role separation — system instructions and user input are mixed in one string.
→ Concept 2 introduces ChatPromptTemplate with typed message roles (system/human/ai).

````python
import sys
sys.path.insert(0, '.')

from PromptUtils import get_llm, TEST_INPUTS, run_inputs
from langchain_core.prompts import PromptTemplate

llm = get_llm()
print(f'LLM ready')
````

### Step 1 — Single Variable Template
The simplest PromptTemplate — one placeholder `{topic}` filled at runtime.

````python
template = PromptTemplate(
    input_variables=['topic'],
    template='Explain {topic} in one sentence.'
)

# Fill the variable to get the final string
prompt_str = template.format(topic='machine learning')
print(f'Formatted prompt: {prompt_str}')

response = llm.invoke(prompt_str)
print(f'Response: {response.content}')
````

### Step 2 — Multiple Variables
Templates can have multiple placeholders. All must be filled before invoking.

````python
multi_template = PromptTemplate(
    input_variables=['topic', 'audience', 'length'],
    template='Explain {topic} to a {audience} in {length}.'
)

prompt_str = multi_template.format(
    topic='neural networks',
    audience='10-year-old',
    length='2 sentences'
)
print(f'Formatted: {prompt_str}\n')

response = llm.invoke(prompt_str)
print(f'Response: {response.content}')
````

### Step 3 — PromptTemplate in an LCEL Chain
Use the `|` pipe to chain: template → llm. The template formats input before it reaches the LLM.

````python
from langchain_core.output_parsers import StrOutputParser

chain_template = PromptTemplate.from_template('Explain {input} in simple terms.')

# LCEL chain: template | llm | parser
chain = chain_template | llm | StrOutputParser()

result = chain.invoke({'input': 'quantum computing'})
print(result)
````

### Full Function
Wrap into a reusable function.

````python
explain_template = PromptTemplate.from_template('Explain {input} clearly and concisely.')
explain_chain    = explain_template | llm | StrOutputParser()

def basic_prompt(query):
    return explain_chain.invoke({'input': query})
````
