```python
from langchain_core.prompts import PromptTemplate
```

```python
#Constructor (PromptTemplate())
prompt1 = PromptTemplate(
    input_variables=["colorful", "socks"],
    template="What is a good name for a company that makes {colorful} {socks}?")

prompt1.format(colorful="colorful", socks="socks")
```

```python
#from_template = PromptTemplate.from_template("What is a good name for a company that makes {product}?")
prompt = PromptTemplate.from_template(
    "What is a good name for a company that makes {product}?")

prompt.format(product="colorful socks")
```

```python
#from_template = PromptTemplate.from_template("What is a good name for a company that makes {product}?")
prompt = PromptTemplate.from_template(
    "What is a good name for a company that makes {product}?")

prompt.format_prompt(product="colorful socks")
```

```python
#from_template = PromptTemplate.from_template("What is a good name for a company that makes {product}?")
prompt = PromptTemplate.from_template(
    "What is a good name for a company that makes {product}?")

prompt.invoke({"product": "colorful socks"})
```

```python
#partial()
prompt = PromptTemplate(
    template="Explain {topic} in {language}",
    input_variables=["topic", "language"]
)

partial_prompt = prompt.partial(
    language="English"
)

print(
    partial_prompt.format(
        topic="Machine Learning"
    )
)
```

```python
#Save the prompt template to a JSON file
prompt.save("prompt_template.json")
```

```python
from langchain_core.prompts import PromptTemplate

sql_prompt = PromptTemplate.from_template(
"""
Convert the following question into SQL.

Schema:
students(
 student_id,
 student_name,
 class,
 total_marks
)

Question:
{question}

Return only SQL. nothing else.
"""
)

query = sql_prompt.format(
    question="Show students in Class 8 who scored more than 80 marks."
)

print(query)
```

```python

from langchain_openai import ChatOpenAI

LLM_MODEL = 'gpt-4o-mini'

def get_llm(temperature=0.0):
    return ChatOpenAI(model=LLM_MODEL, temperature=temperature)
llm = get_llm()
response = llm.invoke(query)
```

```python
response.content
```

## ChatPromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that translates natural language to SQL. return only SQL. nothing else. Schema:students( student_id, student_name, class1, total_marks)"), 
    ("human", "{question}")
])  

prompt = prompt_template.format_prompt(
    question="Show students in Class 8 who scored more than 80 marks."
)
prompt

```

```python
prompt_template = ChatPromptTemplate(
    messages=[
        ("system", "You are a helpful assistant that translates natural language to SQL. return only SQL. nothing else. Schema:students( student_id, student_name, class1, total_marks)"), 
        ("human", "{question}")
    ]
)
prompt = prompt_template.format_prompt(
    question="Show students in Class 8 who scored more than 80 marks."
)
prompt
```

```python
from langchain_openai import ChatOpenAI

LLM_MODEL = 'gpt-4o-mini'

def get_llm(temperature=0.0):
    return ChatOpenAI(model=LLM_MODEL, temperature=temperature)
llm = get_llm()
response = llm.invoke(prompt)
```
