```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts.few_shot import FewShotPromptTemplate
from langchain_core.prompts import FewShotChatMessagePromptTemplate
```

```python
examples = [
    { 
        "input": "List down all students", 
        "output": "select * from students"
    },
    { 
        "input": "List down all students with age greater than 20", 
        "output": "select * from students where age > 20"
    },
    { 
        "input": "List down all students with age greater than 20 and name starting with 'A'", 
        "output": "select * from students where age > 20 and name like 'A%'"
    }
]

example_prompt = ChatPromptTemplate.from_messages([
    ("human", "{input}"),
    ("ai", "{output}")
])  


```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(   
    example_prompt=example_prompt,
    examples=examples
)
    
```

```python
final_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a SQL generator. You will be given a natural language query and you will generate the corresponding SQL query."),
    few_shot_prompt,
    ("human", "{input}")  # human message must be at the end
])

final_prompt.format_messages(input="List down all students with age greater than 20 and name starting with 'A'")
  
```

```python
from langchain_openai import ChatOpenAI

LLM_MODEL = 'gpt-4o-mini'

def get_llm(temperature=0.0):
    return ChatOpenAI(model=LLM_MODEL, temperature=temperature)
llm = get_llm()
response = llm.invoke(final_prompt.format_messages(input="List down all students with age greater than 20 and name starting with 'ac'")  )
print(response.content)
```
